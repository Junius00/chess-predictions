```python
import scipy.stats as st

def plotKDE(x, y):
    #x = data['rating_diff']
    #y = data['turns']

    # Define the borders
    deltaX = (max(x) - min(x))/10
    deltaY = (max(y) - min(y))/10
    xmin = min(x) - deltaX
    xmax = max(x) + deltaX
    ymin = min(y) - deltaY
    ymax = max(y) + deltaY

    # Create meshgrid
    xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]

    # Fit a gaussian kernel using the scipyâ€™s gaussian_kde method
    positions = np.vstack([xx.ravel(), yy.ravel()])
    values = np.vstack([x, y])
    kernel = st.gaussian_kde(values)
    f = np.reshape(kernel(positions).T, xx.shape)
    return xx, yy, f

fig = plt.figure(figsize=(13, 7))
ax = plt.axes(projection='3d')


a, b, c = plotKDE(df3ii['rating_diff'], df3ii['turns'])
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
a, b, c = plotKDE(df3iii['rating_diff'], df3iii['turns'])
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, cmap='terrain', edgecolor='none')
a, b, c = plotKDE(df3i['rating_diff'], df3i['turns'])
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, cmap='ocean', edgecolor='none')

ax.set_xlabel('Rating Difference')
ax.set_ylabel('Turns')
ax.set_title('Surface plot of Gaussian 2D KDE')

fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(0, 85)
```

```python
from mpl_toolkits import mplot3d
from scipy.stats import multivariate_normal
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    index = (np.abs(array - value)).argmin()
    return index
```

```python
#create grid
x = np.linspace(minDiff - 500, maxDiff + 500, 500) #rating diff
y = np.linspace(minTurns - 500, maxTurns + 500, 500) #turns
X, Y = np.meshgrid(x, y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y

#setting parameters
meanInt = find_nearest(data['black_move_3_prob'], data['black_move_3_prob'].mean())
Xcov = np.cov(data['rating_diff'].to_numpy())
Ycov = np.cov(data['turns'].to_numpy())
Zcov = np.cov(data['black_move_1_prob'].to_numpy())


mu = np.array([data['rating_diff'][meanInt], data['turns'][meanInt]]) #mean of the distribution
cov = np.array([[Xcov, 0], [0, Ycov]]) #covariance of the distribution (..?)
rv = multivariate_normal(mu, cov)
Z = rv.pdf(pos)

fig = plt.figure(figsize=(50,50))
plt.rcParams['font.size'] = '30'
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z,cmap='viridis',linewidth=0)
ax.view_init(30, 45)

fig.show()
```
